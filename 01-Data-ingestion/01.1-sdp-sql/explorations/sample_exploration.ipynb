### Ejemplo de Notebook Exploratorio

Utiliza este notebook para explorar los datos generados por el pipeline en tu lenguaje de programación preferido.

**Nota**: Este notebook no se ejecuta como parte del pipeline.

```sql
%python
catalog = "main"
schema = dbName = db = "dbdemos_fsi_fraud_detection"
volume_name = "fraud_raw_data"
```

```sql
%python
sdf = spark.read.json(f"/Volumes/{catalog}/{db}/{volume_name}/transactions")
display(sdf)
# Lee los datos de transacciones en formato JSON y los muestra.
```

```sql
%python
sdf = spark.read.format("csv").option("header", True).option("multiLine", True).load(f"/Volumes/{catalog}/{db}/{volume_name}/customers")
display(sdf)
# Lee los datos de clientes en formato CSV y los muestra.
```

```sql
%python
sdf = spark.read.format("csv").option("header", True).load(f"/Volumes/{catalog}/{db}/{volume_name}/country_code")
display(sdf)
# Lee los datos de códigos de país en formato CSV y los muestra.
```

```sql
%python
sdf = spark.read.format("csv").option("header", True).load(f"/Volumes/{catalog}/{db}/{volume_name}/fraud_report")
display(sdf)
# Lee los datos de reportes de fraude en formato CSV y los muestra.
```
